소문자 처리는 알아서 해줌
중요한 것은 맨 처음이 숫자이며 ,로 단어들이 구분되어야 함

```python
library = {}
cnt = 0
# 단어 사전 및 인덱스값 생성
def check(word):
  global cnt
  if word in library:
    return library[word]
  library[word] = str(cnt)
  cnt += 1
  return library[word]
```

```python
# 텍스트 파일 읽음
def read(path, target):
  file_path = f"{path}/{target}"

  with open(file_path, "r") as f:
    contents = f.read()

  # print(contents)
  base = contents.split("\n")
  print(base)
  for i in range(len(base)):
    if base[i] and base[i][:3] in ["1 -", "2 -", "3 -", "4 -", "5 -", "6 -", "7 -", "8 -", "9 -"]:
      tmp=base[i].replace(",", " ").split()
      for j in range(2, len(tmp)):
        tmp[j] = check(tmp[j].lower())
      print(" ".join(tmp))
      base[i] = " ".join(tmp)
  print(base)
  f.close()

  with open(f'/content/word/T_{target}', 'w') as q:
    for item in base:
      q.write(str(item) + '\n')
  q.close()
```

```python
import os
if not os.path.exists("/content/word"):
  os.makedirs('/content/word')

path = "/content/drive/MyDrive/캡스톤/dataset 백진욱 김하영/클래식 악장 라벨 크롤링 김하영"
for target in os.listdir(path):
  print(target)
  if target[:2] == "E_":
    read(path, target)
```
