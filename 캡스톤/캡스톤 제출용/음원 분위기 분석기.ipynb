```python
#텍스트 파일 읽음
def read(path, target):
  file_path = f"{path}/{target}"

  with open(file_path, "r") as f:
    contents = f.read()

  # print(contents)
  base = contents.split("\n")
  return base
  f.close()
```

```python
#이 path에서 위의 파일을 기준으로 가져와야 함!
#/content/drive/MyDrive/캡스톤/dataset 백진욱 김하영/클래식 악장 음원 크롤링 백진욱/ClassicMusic/Sonata No. 5 C minor, Opus 10/1 (1798)/1
#근데 파일명에 /을 colab에서 가져오니 인식이 안됨!! 따라서 공백으로 변경
import librosa
audio_np=[]
inst=[]
#/content/data/Piano Concerto No.2 In G Minor Op.22 1악장.mp3
import os
from scipy.io import wavfile
import librosa
path = "/content/data"
for file in os.listdir(path):
  if file[-4:] == ".mp3":
    print(f"{file}")
    inst.append([file])
    audio, sr = librosa.load(f"{path}/{file}")
    audio_np.append(audio)
print(len(audio_np), len(inst))
```

```python
#변환 확인
print(len(audio_np), len(inst))
print(len(audio_np[0]), inst[0])
print(audio_np[1][160000:160100])
```

```python
sr = 22050
max_len = 500000
padded_audio_list = []
padded_label_list = []

check  = 0
for i in range(len(audio_np)):
  print(check)
  a = audio_np[i][:max_len]
  if len(a) == max_len:
    padded_audio_list.append(a)
    padded_label_list.append(inst[i])
    copy_cnt = 1
    print(len(a), copy_cnt)
  else:
    padded_audio = librosa.util.fix_length(a, size = max_len)
    padded_audio_list.append(padded_audio)
    padded_label_list.append(inst[i])
    print("처음이 마지막 처리")
    check += 1
    continue
  while len(a) == max_len:
    a = audio_np[i][max_len*copy_cnt:max_len*(copy_cnt+1)]
    if len(a) != max_len and len(a) >= max_len*0.5:
      print(len(a), "나중에 마지막 처리")
      check += 1
      padded_audio = librosa.util.fix_length(a, size = max_len)
      padded_audio_list.append(padded_audio)
      padded_label_list.append(inst[i])
      print(copy_cnt+1)
      break
    elif len(a) != max_len:
      print(len(a), "사라진 마지막 처리")
      check += 1
      break
    padded_audio_list.append(a)
    padded_label_list.append(inst[i])
    copy_cnt += 1
    print(len(a), copy_cnt)

print(padded_audio_list[1][:10], padded_label_list[:10])
print(check, len(padded_audio_list), len(padded_label_list))
```

```python
#에러 체크
for i in range(len(padded_audio_list)):
  if len(padded_audio_list[i]) != max_len:
    print(i, len(padded_audio_list[i]))
    break
```

```python
#0은 없다는 것을 의미하기 위해서 1을 빼서 사전을 분석해야함, 전처리과정에서 0도 라벨값으로 잡아서 1을 더해서 생성함
number = [9, 89, 152, 26, 1, 17, 2, 46, 239, 8, 20, 15, 21, 35, 39, 32, 139, 3, 86, 34, 5, 42, 206, 242, 18, 51, 124, 30, 4, 113, 81, 219, 130, 204, 318, 56, 6, 7, 14, 184, 36, 24, 16, 97, 225, 19, 205, 79, 141] #84 -
```

```python
import xgboost as xgb
result = [[] for _ in range(flattened_data.shape[0])]
model_path = "/content/xgboost/xgb_model"
for class_index in number:
    check_train = xgb.DMatrix(flattened_data)
    
    # 모델 예측
    xgb_model = xgb.Booster()
    xgb_model.load_model(f"{model_path}{class_index}.bin")
    predictions = xgb_model.predict(check_train)
    # print(f"Class {class_index} Predictions:", predictions)
    # print(predictions>0.5)
    for i in range(len(predictions)):
      if predictions[i] > 0.5:
        result[i].append(class_index-1) #단어사전 탐색 처리
print(result)
```

```python
final = [[] for _ in range(len(inst))]
len_list = []
idx = 0
stand = padded_label_list[0]
check = 0
for i in range(len(padded_label_list)):
  check += 1
  if padded_label_list[i] == stand:
    final[idx].append(result[i])
  else:
    len_list.append(check)
    stand = padded_label_list[i]
    idx += 1
    final[idx].append(result[i])
len_list.append(check)
print(check, stand, idx)
print(final)
```

```python
sum_final = [{} for _ in range(len(final))]
for i in range(len(final)):
  # print(len(final[i]))
  for value in final[i]:
    for elem in value:
      if elem in sum_final[i]:
        sum_final[i][elem] += 1
      else:
        sum_final[i][elem] = 1
```

```python
for i in range(len(sum_final)):
  sum_final[i] = dict(sorted(sum_final[i].items(), key=lambda x: x[1], reverse = True))
```

```python
print(len_list)
for i in range(len(len_list)-1, 0, -1):
  len_list[i] = len_list[i]-len_list[i-1]
print(len_list)
for i in range(len(sum_final)):
  print(sum_final[i])
  tmp = []
  for idx, value in sum_final[i].items():
    if value >= len_list[i]//5:
      tmp.append(idx)
  sum_final[i] = tmp
```

```python
def find_word(dictionary, number):
    for word, num in dictionary.items():
        if num == str(number):
            return word
    return None
```
