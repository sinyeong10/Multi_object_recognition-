### 클래식 파일이 있는 사이트 크롤링

```python
url = 'http://www.piano-midi.de/midi_files.htm'
try:
    html = urllib.request.urlopen(url)
except (HTTPError or URLError) as e:
    pass
print(url)
soupClassic = BeautifulSoup(html, 'html.parser')

classicURL = []

tag_table = soupClassic.find('table', attrs={'class':'midi'})
a_tags = tag_table.find_all('a')
a_tags_with_href = [a for a in a_tags if a.get('href')]
for a in a_tags_with_href:
    classicURL.append('http://www.piano-midi.de/'+a['href'])
print(classicURL)
```
