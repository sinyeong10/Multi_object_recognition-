```python
import numpy as np

with open('/content/inst.txt', 'r') as f:
    lines = f.readlines()

inst_list = []

for line in lines:
    row = line.strip().split()  # 공백으로 분리
    base = [0] * 331
    for item in row:
      base[int(item)] = 1#/len(row)  # 정수로 변환
    
    # 열의 길이가 최대 열의 길이와 일치하도록 맞춤
    
    inst_list.append(base)

inst_array = np.array(inst_list)

np.save('/content/inst.npy', inst_array)
```

```python
mfcc_data = np.load('/content/mfcc.npy', allow_pickle=True)
labels = np.load('/content/inst.npy', allow_pickle=True)
```

```python
print(mfcc_data.shape)
print(labels.shape)
```

```python
n_samples = mfcc_data.shape[0]
n_features = mfcc_data.shape[1] * mfcc_data.shape[2]
flattened_data = mfcc_data.reshape(n_samples, n_features)
print(flattened_data.shape)
```

```python
import random

# 같은 인덱스의 서브 리스트들을 추출합니다.
sub_lists = list(zip(flattened_data, labels))

# 추출한 서브 리스트들을 섞습니다.
random.shuffle(sub_lists)

# 섞인 서브 리스트들을 다시 원래 리스트에 삽입합니다.
flattened_data, flattened_labels = zip(*sub_lists)

# 두 개의 리스트의 길이를 구합니다.
length = len(flattened_labels)

# 8:2 비율에 따라 분할할 인덱스를 계산합니다.
split_index = int(length * 0.9)

# 첫 번째 리스트를 분할합니다.
data_train = np.array(flattened_data[:split_index])
data_test = np.array(flattened_data[split_index:])

# 두 번째 리스트를 분할합니다.
labels_train = np.array(flattened_labels[:split_index])
labels_test = np.array(flattened_labels[split_index:])

# 결과를 출력합니다.
print(data_train.shape, data_test.shape)
print(labels_train.shape, labels_test.shape)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.multioutput import MultiOutputClassifier
from sklearn.tree import DecisionTreeClassifier

# 기본 분류기를 생성합니다
base_classifier = DecisionTreeClassifier()

# MultiOutputClassifier를 생성합니다 (기본 분류기를 전달합니다)
multi_output_classifier = MultiOutputClassifier(base_classifier)

# 데이터를 기반으로 MultiOutputClassifier를 학습시킵니다
multi_output_classifier.fit(data_train, labels_train)

predictions = multi_output_classifier.predict(data_test)
print(f"Predictions:", predictions)

#모델 평가
from sklearn.metrics import accuracy_score, log_loss

# 예측 결과와 실제 레이블을 비교하여 정확도 계산
accuracy = accuracy_score(labels_test, predictions)
print(f"Accuracy:", accuracy)

# 로그 손실 계산
logloss = log_loss(labels_test, predictions)
print(f"Log Loss:", logloss)

import joblib
# 모델 저장
joblib.dump(multi_output_classifier, 'multi_output_classifier.pkl')

# # 저장된 모델을 불러옵니다
# loaded_model = joblib.load('multi_output_classifier.pkl')
```

```python
import numpy as np
from sklearn.metrics import hamming_loss, precision_score, recall_score, f1_score

# Hamming Loss 계산
hl = hamming_loss(labels_test, predictions)
print("Hamming Loss:", hl)

# Precision 계산
precision = precision_score(labels_test, predictions, average='micro')
print("Precision:", precision)

# Recall 계산
recall = recall_score(labels_test, predictions, average='micro')
print("Recall:", recall)

# F1-Score 계산
f1 = f1_score(labels_test, predictions, average='micro')
print("F1-Score:", f1)


result = [hl, precision, recall, f1]
```
