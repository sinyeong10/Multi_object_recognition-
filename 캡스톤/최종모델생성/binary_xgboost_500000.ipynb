```python
mfcc_data = np.load('/content/mfcc.npy', allow_pickle=True)
labels = np.load('/content/inst.npy', allow_pickle=True)
```

```python
print(mfcc_data.shape)
print(labels.shape)
```

```python
n_samples = mfcc_data.shape[0]
n_features = mfcc_data.shape[1] * mfcc_data.shape[2]
flattened_data = mfcc_data.reshape(n_samples, n_features)
print(flattened_data.shape)
```

```python
import random

# 같은 인덱스의 서브 리스트들을 추출합니다.
sub_lists = list(zip(flattened_data, labels))

seed_value = 1210
random.seed(seed_value)

# 추출한 서브 리스트들을 섞습니다.
random.shuffle(sub_lists)

# 섞인 서브 리스트들을 다시 원래 리스트에 삽입합니다.
flattened_data, flattened_labels = zip(*sub_lists)

# 두 개의 리스트의 길이를 구합니다.
length = len(flattened_labels)

# 8:2 비율에 따라 분할할 인덱스를 계산합니다.
split_index = int(length * 0.9)

# 첫 번째 리스트를 분할합니다.
data_train = np.array(flattened_data[:split_index])
data_test = np.array(flattened_data[split_index:])

# 두 번째 리스트를 분할합니다.
labels_train = np.array(flattened_labels[:split_index])
labels_test = np.array(flattened_labels[split_index:])

# 결과를 출력합니다.
print(data_train.shape, data_test.shape)
print(labels_train.shape, labels_test.shape)
```

```python
#결과 공간 생성
result = []
```

```python
import numpy as np
import xgboost as xgb
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# One-vs-Rest (OvR)
for class_index in range(labels.shape[1]):
    # 해당 클래스를 양성 클래스로, 나머지 클래스들을 음성 클래스로 변환
    binary_labels = labels_train[:, class_index]
    if not 1 in binary_labels:
      print()
      print(class_index + 1)
      print()
      continue
    # print(binary_labels)
    dtrain = xgb.DMatrix(data_train, label=binary_labels)

    positive_samples = binary_labels.sum()
    negative_samples = len(binary_labels) - positive_samples

    # scale_pos_weight 값 계산
    scale_pos_weight = negative_samples / positive_samples
    
    params = {
        'objective': 'binary:logistic',
        'eval_metric': 'logloss',
        'learning_rate': 0.1,
        'scale_pos_weight': scale_pos_weight
    }

    # XGBoost 모델 학습
    xgb_model = xgb.train(params, dtrain, num_boost_round=10)

    
    binary_labels = labels_test[:, class_index]
    print(binary_labels)
    print(len(binary_labels), sum(binary_labels), len(binary_labels) - sum(binary_labels))
    check_train = xgb.DMatrix(data_test, label=binary_labels)
    # 모델 예측
    predictions = xgb_model.predict(check_train)
    print(f"Class {class_index + 1} Predictions:", predictions)
    print(predictions>0.5)

    #모델 평가
    from sklearn.metrics import accuracy_score, log_loss

    # 예측 결과와 실제 레이블을 비교하여 정확도 계산
    accuracy = accuracy_score(binary_labels, predictions > 0.5)
    print(f"Class {class_index + 1} Accuracy:", accuracy)

    # 로그 손실 계산
    if not 1 in binary_labels:
      print()
      print(class_index + 1)
      print()
    else:
      logloss = log_loss(binary_labels, predictions)
      print(f"Class {class_index + 1} Log Loss:", logloss)

    
    # 오차 행렬 계산
    cm = confusion_matrix(binary_labels, predictions > 0.5)

    print(cm)

    # 정확도 계산
    accuracy = accuracy_score(binary_labels, predictions > 0.5)

    # 정밀도 계산
    precision = precision_score(binary_labels, predictions > 0.5)

    # 재현율 계산
    recall = recall_score(binary_labels, predictions > 0.5)

    # F1 스코어 계산
    f1 = f1_score(binary_labels, predictions > 0.5)

    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)

    result.append([class_index + 1, accuracy, logloss, cm, accuracy, precision, recall, f1])

        # # 모델 저장
    # model_path = f'/content/drive/MyDrive/xgb_model{class_index + 1}.bin'
    # xgb_model.save_model(model_path)
    
    # 모델 저장
    model_path = f'/content/xgb_model{class_index + 1}.bin'
    xgb_model.save_model(model_path)
```
